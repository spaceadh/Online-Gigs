```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
df = pd.read_csv('loan_approval.csv')
```

```python
# Print the number of records in the original dataset
print("Number of records in the original dataset:", len(df))
```

```python
# Code to print the first five lines in each column
df.head()
```

```python
# Code to print the last five lines in each column
df.tail()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Remove records with missing values
df_cleaned = df.dropna()
```

```python
# Print the number of records in the cleaned dataset
print("Number of records in the cleaned dataset:", len(df_cleaned))
```

```python
# Save the cleaned dataset to a new CSV file
df_cleaned.to_csv('loan_approval_cleaned.csv', index=False)
```

```python
# Code to print the first five lines in each column of the new csv file
df_cleaned.head()
```

```python
# Code to print the last five lines in each column of the new csv file
df_cleaned.tail()
```

```python
# Check data types of attributes
data_types = df_cleaned.dtypes
print("Data Types:\n", data_types)
```

```python
# Convert '3+' to integer 3 in 'Dependents' attribute
df_cleaned['Dependents'] = df_cleaned['Dependents'].replace('3+', 3)
```

```python
# Save the cleaned dataset to the same CSV file
df_cleaned.to_csv('loan_approval_cleaned.csv', index=False)
```

```python
# Integer encode categorical attributes
categorical_cols = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']
for col in categorical_cols:
    df_cleaned[col] = df_cleaned[col].astype('category').cat.codes
```

```python
# Save the final preprocessed dataset to a new CSV file
df_cleaned.to_csv('loan_approval_final.csv', index=False)
```
